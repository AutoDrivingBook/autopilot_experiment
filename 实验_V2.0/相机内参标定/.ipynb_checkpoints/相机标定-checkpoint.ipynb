### 4. 1 相机内参构成
相机内参通常由两部分构成：相机矩阵K与畸变系数dist。设空间中的点$P$的坐标为$(x, y, z$，若相机成像没有畸变，则点$P$在图像中的投影可通过相机矩阵计算:
$$\begin{bmatrix}
u^{'} \\
v ^{'}\\
1
\end{bmatrix} = \begin{bmatrix}
f_x & 0 & c_x \\
0 & f_y & c_y\\
0 & 0 & 1
\end{bmatrix} \cdot \begin{bmatrix}
x \\
y\\
z
\end{bmatrix}
$$

其中，$K = \begin{bmatrix}
f_x & 0 & c_x \\
0 & f_y & c_y\\
0 & 0 & 1
\end{bmatrix}$就是相机矩阵。

在现实中，由于透镜的原理，相机成像会产生径向畸变。径向畸变通常使用多项式来建模。设未经畸变的点为$(x, y)$，经畸变修正之后的点为$(x_c, y_c)$，二者之间的关系为：
$$\begin{align}
x_c & = x(1 + k_1 r^2 + k_2 r^4 + k_3 r^6) + 2p_1 xy + p2(r^2 + 2x^2) \\
y_c & = y(1 + k_1 r^2 + k_2 r^4 + k_3 r^6) + p1(r^2 + 2x^2) + 2p_1 xy
\end{align}
$$

其中，$r = \sqrt{x^2 + y^2}$。

在实际中，畸变参数的数量可以通过多项式的增减而变化。通常，都是按照如上的形式，选取五个畸变参数。

#### 2. 提取棋盘内角点
张氏标定法中，通过棋盘内角点来建立图像之间的约束关系，从而得到相机的内参，因此我们需要提取棋盘中的内角点。

```python
# 设置棋盘内角点的size，如(6, 8)表示棋盘的长方向有8列内角点，宽方向有6行内角点，共6*8=48个内角点
pattern_size = (6,8)
# 读取角点
ok, corners = cv2.findChessboardCorners(board_1_gray, pattern_size, None)

# 显示角点
board_1_show = board_1.copy()
if ok:
     for pt in corners:
            point = pt[0]
            cv2.circle(board_1_show, center=(int(point[0]), int(point[1])), radius=5, color=(0, 255, 0), thickness=-1, lineType = cv2.LINE_AA)
            
plt.imshow(board_1_show)
```

在获取上述图像角点之后，要对角点的位置进行进一步的精化，以得到

```python
windows_size = (5, 5)
zeroZone = (-1, -1)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.001)
corners = cv2.cornerSubPix(board_1_gray, corners, windows_size, zeroZone, criteria)

board_1_show_2 = board_1.copy()
for pt in corners:
    point = pt[0]
    cv2.circle(board_1_show_2, center=(int(point[0]), int(point[1])), radius=5, color=(255, 0, 0), thickness=-1, lineType = cv2.LINE_AA)

plt.imshow(board_1_show_2)
```

#### 3. 进行标定
我们通过上述内容学会了如何提取棋盘内角点，当提取出多幅图像的内角点之后，就可以进行标定了

```python
# 提取内角点的函数
def extractChessBoardCorners(image, pattern_size):
    ok, corners = cv2.findChessboardCorners(image, pattern_size, None)
    
    # 如果提取失败，则返回false
    if(ok == False):
        return ok, None
    
    # 提取成功，进行进一步精化
    winSize = (5, 5)
    zeroZone = (-1, -1)
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.001)
    corners = cv2.cornerSubPix(image, corners, winSize, zeroZone, criteria)
    return ok, corners
```

读取多幅图像，提取其中的内角点

```python
img_1 = cv2.imread("data/1_Fisheye.png")
img_2 = cv2.imread("data/2_Fisheye.png")
img_3 = cv2.imread("data/3_Fisheye.png")
img_4 = cv2.imread("data/4_Fisheye.png")
img_5 = cv2.imread("data/5_Fisheye.png")
img_1_gray = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY)
img_2_gray = cv2.cvtColor(img_2, cv2.COLOR_BGR2GRAY)
img_3_gray = cv2.cvtColor(img_3, cv2.COLOR_BGR2GRAY)
img_4_gray = cv2.cvtColor(img_4, cv2.COLOR_BGR2GRAY)
img_5_gray = cv2.cvtColor(img_5, cv2.COLOR_BGR2GRAY)

imgs = [img_1_gray, img_2_gray, img_3_gray, img_4_gray, img_5_gray]

pattern_size = (6,12)
corners_list = []
for i in range(0, len(imgs)):
    ref, c = extractChessBoardCorners(imgs[i], pattern_size)
    if(ref == True):
        corners_list.append(c)

print(len(corners_list))

plt.imshow(img_1)
```

为了进行标定，我们需要先以建立以标定板平面为xOy平面的世界坐标系

```python
# 尺度系数，表示棋盘上的一格代表多现实中多长
scale = 1

# 将角点转换为3d坐标
object_points_list = []
for corners in corners_list:
    object_points = np.zeros((pattern_size[0] * pattern_size[1], 3), np.float32)
    object_points[:, :2] = np.mgrid[0:pattern_size[0],0:pattern_size[1]].T.reshape(-1,2)
    print(len(object_points))
    object_points_list.append(object_points)

# object_points_list = np.array(object_points_list)   
# print(object_points_list)
# corners_list = np.array(corners_list) 
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(object_points_list, corners_list, img_1_gray.shape, None, None)

print(mtx)
print(dist)
```

将图像进行去畸变、打印K矩阵与畸变系数

```python
img_show = cv2.imread("data/1_Fisheye.png")
# newcameramtx, roi=cv2.getOptimalNewCameraMatrix(
#     mtx,dist,
#     (img_1_gray.shape[1], img_1_gray.shape[0]),
#     0,
#     (img_1_gray.shape[1], img_1_gray.shape[0])) # 自由比例参数

dst = cv2.undistort(img_show, mtx, dist, None, None)
plt.imshow(dst)
cv2.imwrite('result.png', dst)
```
